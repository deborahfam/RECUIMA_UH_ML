```python
import pandas as pd
import numpy as np
import data_completion_methods as daco
import data_augmentation_methods as daaug
import util
PATH_SAVE = 'augmented_data/'
PATH_AUTOCOMPLETE = 'autocomplete_data/'
PATH_DEFAULT = 'default_values/'
```

```python
original_data = pd.read_csv(f'{PATH_AUTOCOMPLETE}completed_data__C01__01.csv', index_col=0)
dieds = original_data[original_data['estado_vital'] == 1]
dieds
```

```python
new_dieds = daaug.create_many_mod_record_from_json(1000,dieds,'augmented_data/feature_augmented__01.json')
new_dieds.reset_index(0,drop=True, inplace=True)
new_dieds
```
